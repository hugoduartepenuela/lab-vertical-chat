```python
#if you need an API Key from OpenAI
#https://platform.openai.com/account/api-keys

from openai import OpenAI
import os
from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv())

OPENAI_API_KEY  = os.getenv('OPENAI_API_KEY')
```

```python
client = OpenAI(
    # This is the default and can be omitted
    api_key=OPENAI_API_KEY,
)

def continue_conversation(messages, temperature=0):
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=messages,
        temperature=temperature,
    )
    #print(str(response.choices[0].message["content"]))
    return response.choices[0].message.content
```

```python
def add_prompts_conversation(_):
    #Get the value introduced by the user
    prompt = client_prompt.value_input
    client_prompt.value = ''

    #Append to the context the User prompt.
    context.append({'role':'user', 'content':f"{prompt}"})

    #Get the response.
    response = continue_conversation(context)

    #Add the response to the context.
    context.append({'role':'assistant', 'content':f"{response}"})

    #Update the panels to show the conversation.
    panels.append(
        pn.Row('User:', pn.pane.Markdown(prompt, width=600)))
    panels.append(
        pn.Row('Assistant:', pn.pane.Markdown(response, width=600)))

    return pn.Column(*panels)
```

```python
import panel as pn

pn.extension()

context = [{
    'role': 'system',
    'content': """
Act as a Pokemon Match Bot.

Your task is to help the user discover which Pokemon they would probably like the most based on their personality, style, interests, and preferences.

Start by greeting the user in a fun and friendly tone.
Then ask one question at a time to learn about:
- their favorite colors
- animals they like
- preferred vibe or aesthetic
- favorite environments
- whether they like cute, strong, mysterious, elegant, funny, or dark characters
- favorite Pokemon types or elemental powers
- whether they value design, strength, rarity, nostalgia, or emotional connection more
- whether they prefer classic generations or newer Pokemon

Once you have enough information, choose the Pokemon that matches them best.
Then explain why in a short and clear way.
Also suggest 2 alternative Pokemon they may also like.

Keep the conversation short, fun, and engaging.
Do not answer randomly.
Always connect the final recommendation to the user's answers.
If the information is limited, make a reasonable guess and say so.
"""
}]

panels = []

client_prompt = pn.widgets.TextInput(value="Hi", placeholder="Enter text here…")
button_conversation = pn.widgets.Button(name="Talk")

def fake_llm_response(user_message):
    msg = user_message.lower()

    if "hi" in msg or "hello" in msg:
        return "Hey! Let’s find your Pokemon match. First: what colors do you usually like the most?"
    
    elif any(color in msg for color in ["blue", "red", "black", "white", "purple", "green", "yellow", "pink"]):
        return "Nice. Now tell me: do you usually prefer cute, powerful, mysterious, elegant, funny, or darker characters?"
    
    elif any(word in msg for word in ["cute", "powerful", "strong", "mysterious", "elegant", "funny", "dark"]):
        return "Good choice. What kind of environment do you like most: ocean, forest, mountains, city, night, or space?"
    
    elif any(place in msg for place in ["ocean", "forest", "mountains", "city", "night", "space", "sea"]):
        return "Got it. Last thing: do you care more about design, strength, rarity, nostalgia, or emotional connection?"
    
    elif any(value in msg for value in ["design", "strength", "rarity", "nostalgia", "emotional"]):
        return """Your best Pokemon match is **Umbreon**.

Why? You seem to like a sleek, mysterious, stylish vibe with strong visual identity and emotional appeal. Umbreon fits that perfectly.

Two other Pokemon you may also like:
- **Lucario** — for a cool, strong, balanced energy
- **Greninja** — for a modern, sharp, elegant style"""
    
    else:
        return "Interesting. Tell me a bit more about your style, favorite vibe, or the kind of Pokemon you usually notice first."

def add_prompts_conversation(event):
    prompt = client_prompt.value
    response = fake_llm_response(prompt)

    panels.append(
        pn.Row("User:", pn.pane.Markdown(prompt, width=600))
    )
    panels.append(
        pn.Row("Bot:", pn.pane.Markdown(response, width=600))
    )

    client_prompt.value = ""
    return pn.Column(*panels)

interactive_conversation = pn.bind(add_prompts_conversation, button_conversation)

dashboard = pn.Column(
    client_prompt,
    pn.Row(button_conversation),
    pn.panel(interactive_conversation, loading_indicator=True),
)

dashboard
```

## Final Report

In this exercise, I designed different chatbots by modifying the system prompt and adapting the conversation flow depending on the objective. The three main use cases were: an investment fund assistant, a humanitarian recruitment assistant, and a Pokémon preference bot.

The key idea was to keep a similar structure across all bots: greeting the user, asking questions step by step, collecting relevant information, summarizing it, and suggesting a final action or result. However, each use case required a different tone and level of responsibility.

For the investment fund chatbot, the main challenge was avoiding unrealistic or risky responses, such as promising returns or giving financial advice. The prompt had to clearly restrict the model and keep the tone professional. This made the bot more credible but also more limited in its responses.

In the humanitarian recruitment chatbot, the focus was on collecting candidate information in a structured and respectful way. The difficulty here was balancing empathy with realism, especially when describing working conditions, without making promises about safety or selection.

The Pokémon chatbot was the most flexible and creative. It allowed a more playful tone and showed how the model can personalize responses based on user preferences. This version worked better in terms of engagement and felt more natural in conversation.

Some variations did not work well. When the prompt was too vague, the model gave generic or random answers. In some cases, if the instructions were not strict enough, the chatbot skipped steps or did not ask enough questions before giving a final answer. Also, using a simple rule-based simulation (fake responses) made the chatbot predictable and repetitive.

Overall, the results show that prompt design strongly affects the quality, structure, and reliability of the chatbot.

## What I Learned

I learned that the structure of the prompt is critical. If the instructions are not clear, the chatbot quickly becomes inconsistent or gives low-quality answers.

I also realized that different contexts require different levels of control. For sensitive topics like finance or recruitment, the prompt must include restrictions to avoid incorrect or risky responses. In contrast, for creative tasks like the Pokémon bot, the model performs better and feels more natural.

Another important point is that asking questions step by step improves the interaction. If the chatbot tries to do everything at once, the result is worse.

Finally, I noticed that simple simulated responses are useful for testing the interface, but they do not reflect the real capabilities of a language model. To build a better chatbot, it is necessary to connect it to an actual model and handle conversation memory properly.
